# element wise multiplicaton

## a * b
## np.multiply(a,b)

> Broadcasts dimensions.
Multiplies corresponding elements.

# matrix multiplication
## a @ b
## np.matmul(a,b)
> dot products
row × column rules

# outer product
## np.outer(a,b)
> Computes ALL pairwise multiplications.
```python
Equivalent mathematically to:

a.reshape(-1,1) * b.reshape(1,-1)
```
